```python
def get_portfolio_returns(tickers, start_date, end_date):
      
    portfolio_returns = pd.DataFrame()
    for i, ticker in enumerate(tickers):
        df = pro.daily(ts_code = ticker, start_date = start_date, end_date = end_date)
        df = df.set_index('trade_date')
        portfolio_returns[ticker] = df['pct_chg'] / 100
    portfolio_returns.index = pd.to_datetime(portfolio_returns.index)
    portfolio_returns = portfolio_returns.sort_values(['trade_date'])
    
    return portfolio_returns
```
